```python
from jira import JIRA
import logging
import os

logFormatter = logging.Formatter("%(asctime)s [%(threadName)-12.12s] [%(levelname)-5.5s]  %(message)s")
logger = logging.getLogger()
logPath = os.path.join(os.curdir, "logs")
fileName = "jiraextract"

fileHandler = logging.FileHandler("{0}/{1}.log".format(logPath, fileName))
fileHandler.setFormatter(logFormatter)
logger.addHandler(fileHandler)

consoleHandler = logging.StreamHandler()
consoleHandler.setFormatter(logFormatter)
logger.addHandler(consoleHandler)
```

```python
jira = JIRA('https://jira.spring.io')

block_size = 1000
block_num = 0
allissues = []
while True:
     start_idx = block_num*block_size
     issues = jira.search_issues('project=XD', start_idx, block_size)
     if len(issues) == 0:
         break
     block_num += 1
     for issue in issues:
         print('%s: %s' % (issue.key, issue.fields.summary))
         allissues.append(issue)
```

```python
len(allissues)
```

```python
import pandas as pd
issues = pd.DataFrame()

for issue in allissues:
    issue_dict = {
        'key': issue.key,
        'assignee': issue.fields.assignee,
        'creator': issue.fields.creator,
        'reporter': issue.fields.reporter,
        'description': issue.fields.description,
        'summary': issue.fields.summary
    }

    issues = issues.append(issue_dict, ignore_index= True)

issues.head()
```

```python
issues.to_csv('issues-xd.csv', encoding='utf-8', header=True, index=True)
```

```python
from fuzzywuzzy import fuzz

issues["accuracy"] = issues.apply(lambda row: fuzz.token_sort_ratio(row['summary'],'Timezone not working for Admin config.'), axis=1)
issues.head()
```

```python
issues = issues.sort_values("accuracy", ascending=False)
issues.head()
```

```python
print("best matching issue: {} \naccuracy: {}%".format(issues['summary'].iloc[0], issues['accuracy'].iloc[0]))
```

```python
from fuzzywuzzy import process

process.extract('Timezone not working for Admin config.', issues['summary'])
```
